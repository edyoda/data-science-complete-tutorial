#### Thumb-Rules of handling missing data
* Drop column with more than 80% missing data
* Drop rows which have very less number of missing data
* Otherwise try some missing data filling techniques

```python
titanic_data.dropna(inplace=True)
```

```python
titanic_data.info()
```

```python
#less than 6 non-null values will be dropped
movie_data.dropna(thresh=6)
```

```python
#drop column based on configured threshold
movie_data.dropna(thresh=5,axis=1)
```

```python
movie_data.drop(['Adam Cohen'],axis=1, inplace=True)
```

```python
movie_data.drop(['Scarface'], inplace=True)
```

### Duplicate Data Handling

```python
df = pd.DataFrame({'A':[1,2,3,2,5], 'B':[1,2,5,2,2], 'C':[2,2,2,2,2]})
```

```python
df[df.duplicated()]
```

```python
df
```

```python
df.duplicated(subset=['B','C'])
```

```python
df.drop_duplicates(subset=['B','C'])
```
