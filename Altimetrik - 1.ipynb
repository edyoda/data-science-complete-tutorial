```python
def f(d):
    if (d == 'sales') | (d == 'marketing'):
        return 'salesAndMarketing'
    else:
        return d
    
hr_data['new_dept'] = hr_data.department.map(f)
```

```python
hr_data.sample(10)
```

```python
def f(d):
    return str(int(d/24)) + ' days & ' + str(d%24) + ' hours'
    
hr_data['new_avg_monthly_hrs'] = hr_data.average_montly_hours.map(f)
```

```python
hr_data.head()
```

```python
def f(r):
    if r.number_project >= 4:
        return 'high'
    else:
        return r.salary
hr_data['new_salary'] = hr_data.apply(f, axis=1)
```

```python
hr_data[hr_data.new_salary != hr_data.salary]
```
